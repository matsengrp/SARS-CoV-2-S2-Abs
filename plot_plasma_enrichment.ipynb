```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import warnings
warnings.filterwarnings('ignore')
```

```python
plasma_C20_df = pd.read_csv('data/C20_plasma_enrichment.csv',index_col=0)
plasma_C68_df = pd.read_csv('data/C68_plasma_enrichment.csv',index_col=0)
```

```python
oi_blue       = '#0072B2'
oi_vermillion = '#D55E00'

peptide_locs = plasma_C20_df['Loc'].to_numpy()
enrC20 = plasma_C20_df['enrichment'].to_numpy()
enrC68 = plasma_C68_df['enrichment'].to_numpy()

fig = plt.figure(figsize=[15,8])
plt.plot(peptide_locs[685:], enrC20[685:], linewidth=3, alpha=0.8, color=oi_blue,       label='C20 plasma')
plt.plot(peptide_locs[685:], enrC68[685:], linewidth=3, alpha=0.8, color=oi_vermillion, label='C68 plasma')
plt.grid()
plt.xticks(fontsize=24)
plt.xlabel('peptide location', fontsize=30, labelpad=20)
plt.yticks(fontsize=24)
plt.ylabel('enrichment', fontsize=30, labelpad=20)
plt.legend(fontsize=30)
plt.tight_layout()
plt.savefig(f'plasma_20C_vs_268C_spike_S2.pdf')
plt.show()
```
