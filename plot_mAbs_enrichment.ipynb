```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import warnings
warnings.filterwarnings('ignore')
```

```python
oi_blue       = '#0072B2'
oi_vermillion = '#D55E00'
    
for mAb in ['CV3-25','1A9']:
    df = pd.read_csv(f'data/{mAb}_mAb_enrichment.csv', index_col=0)

    peptide_locs = df['Loc'].to_numpy()
    enr0 = df['enr0'].to_numpy()
    enr1 = df['enr1'].to_numpy()
    
    fig = plt.figure(figsize=[15,8])
    plt.plot(peptide_locs, enr0, linewidth=1, alpha=0.7, color=oi_blue)
    plt.plot(peptide_locs, enr1, linewidth=1, alpha=0.7, color=oi_vermillion)
    plt.grid()
    plt.title(f'{mAb}',fontsize=30)
    plt.xticks(fontsize=24)
    plt.xlabel('peptide location', fontsize=30, labelpad=20)
    plt.yticks(fontsize=24)
    plt.ylabel('enrichment', fontsize=30, labelpad=20)
    plt.tight_layout()
    plt.savefig(f'{mAb}_mAb_enrichment.pdf')
    plt.show()
```
